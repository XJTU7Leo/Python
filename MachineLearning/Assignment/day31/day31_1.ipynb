# 折线图

```python
x = range(1,8) # x轴的位置
y = [17, 17, 18, 15, 11, 11, 13]
# 传入x和y, 通过plot画折线图
plt.plot(x,y)#x可以是迭代器，生成器，列表，元组
plt.show()
```

```python
x = range(1,8) # x轴的位置
y = [17, 17, 18, 15, 11, 11, 13]
plt.figure(figsize=(20,8),dpi=80)
# 颜色为红色，线宽为2.0，透明度0.5
plt.plot(x,y,color='red',linewidth=2.0,linestyle='-.',alpha=0.5,marker='o')

plt.savefig('./t1.svg')
plt.show()
```

```python
import random
x = range(2,26,2) # x轴的位置
y = [random.randint(15, 30) for i in x]
plt.figure(figsize=(20,8),dpi=80)

# 构造x轴刻度标签
x_ticks=range(1,25)
x_ticks_label = ["{}:00".format(i) for i in x_ticks]

#rotation = 45, 旋转45度
plt.xticks(x_ticks,x_ticks_label,rotation=45)

# 设置y轴的刻度标签
y_ticks_label = ["{}℃".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_ticks_label)

plt.plot(x,y)
```

# 散点图

```python
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17, 18, 21, 16, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 22,
     23]
x = range( 1, 32 )

# 设置图形大小
plt.figure( figsize=(20, 8), dpi=80 )
# 使用scatter绘制散点图
plt.scatter( x, y, label='3月份' )
# 调整x轴的刻度
my_font = font_manager.FontProperties( fname='/System/Library/Fonts/PingFang.ttc', size=18 )

_xticks_labels = ['3月{}日'.format( i ) for i in x]

plt.xticks( x[::3], _xticks_labels[::3], fontproperties=my_font, rotation=45)
plt.xlabel( ' 日 期 ', fontproperties=my_font )
plt.ylabel( '温度', fontproperties=my_font )
# 图 例
plt.legend( prop=my_font )
plt.show()
```

# 柱状图

```python
my_font =  font_manager.FontProperties(fname='/System/Library/Fonts/PingFang.ttc',size=16)
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]

plt.figure(figsize=(20,8),dpi=80)

# 绘制条形图的方法
'''
width=0.3  条形的宽度
'''
rects = plt.bar(range(len(b)),b,width=0.3,color='r')
plt.xticks(range(len(a)),a,fontproperties=my_font,rotation=45)
for rect in rects: # 遍历每一个柱形
    height = rect.get_height() # 获取柱形高度
    plt.text(rect.get_x()+rect.get_width()/2,height+0.4,str(height),ha='center',va='bottom')
plt.show()
```

# 饼状图

```python
import matplotlib
label_list = ["第一部分", "第二部分", "第三部分"]  # 各部分标签
size = [55, 35, 10]    # 各部分大小
color = ["red", "green", "blue"]   # 各部分颜色
explode = [0, 0.05, 0] # 各部分突出值
"""
绘 制 饼 图
explode： 设 置 各 部 分 突 出
label: 设 置 各 部 分 标 签
labeldistance:设置标签文本距圆心位置，1.1表示1.1倍半径
autopct： 设 置 圆 里 面 文 本
shadow：设置是否有阴影
startangle：起始角度，默认从0开始逆时针转
pctdistance：设置圆内文本距圆心距离返回值
l_text：圆内部文本，
matplotlib.text.Text object
p_text：圆外部文本
"""
matplotlib.rcParams['font.sans-serif']=['Songti SC']
patches, l_text, p_text = plt.pie(size,explode=explode, colors=color, labels=label_list,
                                  labeldistance=1.1, autopct="%1.1f%%", shadow=True,
                                  startangle=90, pctdistance=0.6)
plt.axis("equal")  # 设置横轴和纵轴大小相等，这样饼才是圆的
plt.legend()
plt.show()
```
